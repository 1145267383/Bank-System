```python
import sqlite3 as sq
import pandas as pd
import numpy as np
import getpass as gp
#Enter Number.
#f0>>Abave From f0 (int).
#f1>>Ander From f1 (int).
#e>>Print User(str).
#u>>User Enter(str).

def intif(f0 ,f1 ,u ,e ):
    while True:
        #This Number.
        try:
            i = int(input(f"Enter Your {u}: "))
            #Abave from f0 and ander f1.
            if (i>=f0) and (i<=f1) :
                return i
                break
            else:
                print(e)
                continue

        except:
            print("This not Nunber")
            continue

################################
#Enter String number.
def strif(f0 ,f1 ,u ,e ):
    while True:
        s = input(f"Enter Your {u}: ")
        # this Numder.
        try:
            int(s)
            # len between [f0:f1].   
            if (f0<=len(s)<=f1):
                return s
                break
            else:
                print(e)
                continue

        except:
            print("This not Nunber")
            continue
            
#####################
#Save Data
def savedata(tabl,vu,data):
    con = sq.connect("Users.db")
    sql = f"INSERT INTO {tabl} VALUES ({vu})"
    cur = con.execute(sql,data)
    con.commit()
    con.close()

    
#####################
#Read Data.
def readtabl(tabl,colm):
    con = sq.connect("Users.db")
    df =pd.read_sql_query (f"SELECT {colm} FROM {tabl}",con)
    return df
    con.close()

    
####################
#chek Accont 
def chek(ph,pasd):
    ip=str(ph[-7::-1]+pasd[-2::-1])
    try:
        key=str(readtabl(f"ip WHERE ID={ip} And Password={pasd}","ID")["ID"][0])
        print("Log in your account successfuly!")
        return key
    except:
        print("you not have account")
        return 0

###################
#Delete Account.
def delet_acount(tabl,f):
    con = sq.connect("Users.db")
    sql = f"DELETE FROM {tabl} WHERE {f}"
    cur = con.execute(sql)
    
    con.commit()
    con.close()
    

#####################
#updata.
def updata(key,v_amount,k_mt):
    bdf = readtabl(f"balance WHERE ID={key}","Diposte,Withdraw,Balance")
    amot = {"1":"Diposte","2":"Withdraw"}
    k=amot[k_mt]
    amount=int(v_amount+bdf[k][0])
    if k_mt=="1":
        balan=int(bdf["Balance"][0]+v_amount)
    elif k_mt=="2":
        balan=int(bdf["Balance"][0]-v_amount)
        
    con = sq.connect("Users.db")
    sql = f"UPDATE balance SET {k}=?,Balance=? WHERE ID=?"
    data = [amount,balan,key]
    cur=con.execute(sql,data)
    
    con.commit()
    con.close()
    return balan

################################
# create new account .
class User : 
    def __init__ (self ,key ) : 
        self.df_us=readtabl(f"user WHERE ID={key}","*")
        self.name = self.df_us["Name"][0] 
        self._phone = self.df_us["Phone"][0]
        self.age = self.df_us["Age"][0]
        self.gender = self.df_us["Gander"][0]
        
    def show_details(self) : 
        print ("User Detalis " ) 
        print ("Name: " , self.name) 
        print (f"Phone: {self._phone}")
        print (f'Age: {self.age}') 
        print ('Gender: {}'.format(self.gender))
                    
        ##############################################################      
# user account
class Bank (User) : 
    def __init__ (self ,key) : 
        User.__init__(self ,key)
        self.balance = readtabl(f"balance WHERE ID={key}","Balance")["Balance"][0]
    def diposte (self,amount) :
        self.balance = updata(key,amount,"1") 
        return self.balance 
    
    def withdraw (self , amount) : 
        self.balance =  updata(key,amount,"2") 
        return self.balance 
        
    def view_balance (self) : 
        return self.balance
        
      
```
